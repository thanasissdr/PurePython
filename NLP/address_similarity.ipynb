```python
# https://en.wikipedia.org/wiki/Street_suffix
with open('./src/suffixes/address/us.txt', 'r') as f:
    data = f.readlines()
    stopwords = []
    for i in data:
        tokenized = re.split(' ', i)
        for token in tokenized:
            stopwords.append(token.lower().strip())
            
```
